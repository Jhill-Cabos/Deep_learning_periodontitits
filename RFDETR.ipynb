```python
from rfdetr import RFDETRSmall
import torch, os

DATASET_DIR = "/content/drive/MyDrive/EDet"   # <- your COCO folder
OUTPUT_DIR  = "/content/drive/MyDrive/runs/RFDETR_OCT15"
os.makedirs(OUTPUT_DIR, exist_ok=True)
```

```python
model = RFDETRSmall()  # or RFDETRSmall(), RFDETRMedium() when you want bigger accuracy

model.train(
    dataset_dir=DATASET_DIR,
    output_dir=OUTPUT_DIR,
    epochs=150,                 # your requested budget
    batch_size=4,               # per-GPU micro-batch (T4 memory friendly)
    grad_accum_steps=4,         # 4*4 = 16 effective batch
    lr=1e-4,                    # good starting LR
    weight_decay=1e-4,
    resolution=512,             # try 576 if VRAM allows
    device="cuda" if torch.cuda.is_available() else "cpu",
    early_stopping=True,        # stops if mAP plateaus
    early_stopping_patience=55, # tweak as you like
    tensorboard=True            # optional logging
)
```
