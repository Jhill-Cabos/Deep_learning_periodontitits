```python
from ultralytics import YOLO

model = YOLO("yolo11s.pt")

results = model.train(
    data="/content/drive/MyDrive/Dataset/data.yaml",
    epochs=150,
    batch=16,
    workers=2,
    device=0,
    patience=30,
    save=True,
    amp=False,
    pretrained=True,
    lr0=0.0005,
    lrf=0.01,
    momentum=0.937,
    weight_decay=0.0005,
    warmup_epochs=3,
    box=7.5,
    cls=0.5,
    hsv_h=0.015,
    hsv_s=0.5,
    hsv_v=0.4,
    degrees=0.0,
    translate=0.05,
    scale=0.5,
    fliplr=0.5,
    mosaic=0.8,
    mixup=0.0
)
```
